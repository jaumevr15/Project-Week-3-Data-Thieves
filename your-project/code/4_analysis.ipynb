```python
#To import the necessary libraries.

import json
import requests
import pandas as pd
import numpy as np
import random
```

```python
#To read the combined csv with all the video from the three websites to perform some analysis.

data = pd.read_csv("1_videos_combined_dataset.csv", encoding='utf8', sep=";", index_col = 0)
```

```python
# To show the header and make sure that the data was read correctly.

data.head()
```

```python
# To drop the unnamed column at the end of the table

data.drop(columns=['Unnamed: 8'], inplace = True)
```

```python
# Mean number of views for all the videos, from the three websites combined

data.Number_of_Views.mean()
```

```python
# Mean number of views from Pornhub website

Pornhub = data.loc[data["Website"] == "Pornhub"]
Pornhub.Number_of_Views.mean()
```

```python
# Mean number of views from Xhamster website

Xhamster = data.loc[data["Website"] == "Xhamster"]
Xhamster.Number_of_Views.mean()
```

```python
# Mean number of views from Redtube website

Redtube = data.loc[data["Website"] == "Redtube"]
Redtube.Number_of_Views.mean()
```

```python
# To get the top 20 videos based on number of views

Top_20_videos = data.nlargest(20, 'Number_of_Views')
```
